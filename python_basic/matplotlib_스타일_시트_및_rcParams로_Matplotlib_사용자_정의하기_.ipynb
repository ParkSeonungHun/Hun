# 스타일 시트와 rcParams로 Matplotlib 사용자 정의하기

## 스타일 시트 사용하기

이 style패키지는 matplotlib rc 파일 (시작시 Matplotlib 구성을 위해 읽음) 과 동일한 매개 변수를 사용하여 쉽게 전환 할 수있는 플로팅 "스타일"에 대한 지원을 추가합니다 .

Matplotlib에서 제공하는 미리 정의 된 여러 스타일이 있습니다. 예를 들어 ggplot의 미학을 에뮬레이트하는 "ggplot"이라는 미리 정의 된 스타일이 있습니다 ( R에 대한 인기있는 플로팅 패키지 ). 이 스타일을 사용하려면 다음을 추가하십시오.

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
plt.style.use('ggplot')
data = np.random.randn(50)
```

사용 가능한 모든 스타일을 나열하려면 다음을 사용하십시오.

```
print(plt.style.available)
```

mpl_configdir/stylelib스타일의 이름이 같으면 의 사용자 정의 스타일 시트 가 Matplotlib에서 정의한 스타일 시트를 재정의합니다.

# Matplotlib rcParams

## 동적 RC 설정

python 스크립트에서 또는 python 셸에서 대화식으로 기본 rc 설정을 동적으로 변경할 수도 있습니다. 모든 rc 설정은 matplotlib.rcParamsmatplotlib 패키지에 대한 전역 인 이라는 사전과 유사한 변수에 저장 됩니다. rcParams는 다음과 같이 직접 수정할 수 있습니다.

```
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '--'
plt.plot(data)
```

참고 일반적인 변경하기 위해 것을 plot색상을 사용자가 변경해야 prop_cycle의 재산 축 :

```
mpl.rcParams['axes.prop_cycle'] = cycler(color=['r', 'g', 'b', 'y'])
plt.plot(data)  # first color is red
```

Matplotlib는 rc 설정을 수정하기위한 몇 가지 편리한 기능도 제공합니다. matplotlib.rc키워드 인수를 사용하여 한 번에 단일 그룹의 여러 설정을 수정하는 데 사용할 수 있습니다.

```
mpl.rc('lines', linewidth=4, linestyle='-.')
plt.plot(data)
```
