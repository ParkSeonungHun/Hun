```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

# 통계적 관계 시각화

통계 분석은 데이터 세트의 변수가 서로 어떻게 관련되어 있는지 그리고 이러한 관계가 다른 변수에 어떻게 의존하는지 이해하는 과정입니다. 데이터가 제대로 시각화되면 인간의 시각 시스템이 관계를 나타내는 추세와 패턴을 볼 수 있기 때문에 시각화는이 프로세스의 핵심 구성 요소가 될 수 있습니다.

이 튜토리얼에서는 세 가지 seaborn 함수에 대해 설명합니다. 우리가 가장 많이 사용할 것은 relplot(). 이것은 두 가지 일반적인 접근 방식 인 산점도 및 선 플롯을 사용하여 통계적 관계를 시각화 하는 그림 수준 함수 입니다. relplot()a FacetGrid를 두 축 수준 함수 중 하나와 결합 합니다.

scatterplot()(사용 kind="scatter"; 기본값)

lineplot()(와 함께 kind="line")

앞으로 보게 될 이러한 함수는 복잡한 데이터 세트 구조를 표현할 수있는 간단하고 이해하기 쉬운 데이터 표현을 사용하기 때문에 상당히 밝을 수 있습니다. 색조, 크기 및 스타일의 의미를 사용하여 최대 3 개의 추가 변수를 매핑하여 향상시킬 수있는 2 차원 그래픽을 플로팅하기 때문에 그렇게 할 수 있습니다.

## 산점도와 변수 연결하기

산점도는 통계 시각화의 주요 요소입니다. 점 구름을 사용하여 두 변수의 공동 분포를 묘사하며, 여기서 각 점은 데이터 세트의 관측치를 나타냅니다. 이 묘사를 통해 눈은 그들 사이에 의미있는 관계가 있는지에 대한 상당한 양의 정보를 추론 할 수 있습니다.

seaborn에서 산점도를 그리는 방법에는 여러 가지가 있습니다. 두 변수가 모두 숫자 일 때 사용해야하는 가장 기본적인 것은 scatterplot()함수입니다. 에서 범주 시각화 튜토리얼 , 우리는 범주 데이터를 시각화하기 위해 산점도를 사용하기위한 전문 도구를 볼 수 있습니다. 의 scatterplot()기본값 kind입니다 relplot()(을 설정하여 강제 할 수도 있습니다 kind="scatter").

```
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill",y="tip",data=tips);
```

점이 2 차원으로 플로팅되는 동안 세 번째 변수에 따라 점을 채색하여 다른 차원을 플롯에 추가 할 수 있습니다. seaborn에서는 포인트의 색상이 의미를 얻기 때문에 "색조 의미"를 사용한다고합니다.

```
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips);
```

클래스 간의 차이를 강조하고 접근성을 높이기 위해 각 클래스에 대해 다른 마커 스타일을 사용할 수 있습니다.

```
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

각 포인트의 색조와 스타일을 개별적으로 변경하여 4 개의 변수를 표현할 수도 있습니다. 그러나 눈은 색보다 모양에 훨씬 덜 민감하기 때문에 신중하게 수행해야합니다.

```
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

위의 예에서 색조 의미 체계는 범주 형이므로 기본 품질 팔레트 가 적용되었습니다. 색조 의미가 숫자 인 경우 (특히 부동으로 캐스팅 할 수있는 경우) 기본 색상은 순차적 팔레트로 전환됩니다.

```
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

두 경우 모두 색상 팔레트를 사용자 정의 할 수 있습니다. 그렇게하기위한 많은 옵션이 있습니다. 여기에서 문자열 인터페이스를 사용하여 순차 팔레트를 사용자 정의합니다 cubehelix_palette().

```
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

세 번째 종류의 의미 변수는 각 포인트의 크기를 변경합니다.

```
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

와 달리 matplotlib.pyplot.scatter()변수의 리터럴 값은 점의 영역을 선택하는 데 사용되지 않습니다. 대신 데이터 단위의 값 범위가 영역 단위의 범위로 정규화됩니다. 이 범위는 사용자 정의 할 수 있습니다.

```
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

통계적 관계를 표시하기 위해 다른 의미 체계를 사용하는 방법을 사용자 지정하는 더 많은 예제가 scatterplot()API 예제에 나와 있습니다.



## 라인 플롯으로 연속성 강조

산점도는 매우 효과적이지만 보편적으로 최적화 된 시각화 유형은 없습니다. 대신, 시각적 표현은 데이터 셋의 세부 사항과 플롯으로 답하려는 질문에 맞게 조정되어야합니다.

일부 데이터 세트의 경우 한 변수의 변화를 시간 함수 또는 유사한 연속 변수로 이해하려고 할 수 있습니다. 이 상황에서 좋은 선택은 선 그림을 그리는 것입니다. seaborn에서이 lineplot()기능은 직접 또는 다음 relplot()을 설정 하여 수행 할 수 있습니다 kind="line".

```
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

때문에 lineplot()가장 자주 그리기 위해 노력하고 있다고 가정 y의 함수로 x, 기본 동작은으로 데이터를 정렬 할 수있다 x하려하기 전에 값을. 그러나 다음과 같이 비활성화 할 수 있습니다.

```
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

## 집계 및 불확실성 표현

더 복잡한 데이터 세트는 동일한 x변수 값에 대해 여러 측정 값을 갖습니다 . seaborn의 기본 동작 x은 평균과 평균 주위에 95 % 신뢰 구간을 플로팅하여 각 값 에서 여러 측정 값 을 집계 하는 것입니다.

```
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

신뢰 구간은 부트 스트랩을 사용하여 계산되며, 대규모 데이터 세트의 경우 시간이 많이 소요될 수 있습니다. 따라서 비활성화 할 수 있습니다.

```
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

특히 더 큰 데이터의 경우 또 다른 좋은 옵션은 신뢰 구간 대신 표준 편차를 플로팅하여 각 시점에서 분포의 산포를 나타내는 것입니다.

```
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

집계를 모두 끄려면 데이터에 각 지점에서 여러 관측치가있는 경우 estimator매개 변수를 None이로 설정하면 이상한 효과가 발생할 수 있습니다.

```
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
```

## 시맨틱 매핑을 사용하여 데이터의 하위 집합 플로팅

이 lineplot()함수는 다음과 같은 유연성을 갖 scatterplot()습니다. 플롯 요소의 색조, 크기 및 스타일을 수정하여 최대 3 개의 추가 변수를 표시 할 수 있습니다. 이는와 동일한 API를 사용하므로 scatterplot()matplotlib에서 선과 점의 모양을 제어하는 ​​매개 변수에 대해 생각할 필요가 없습니다.

의미 체계를 사용 lineplot()하면 데이터가 집계되는 방식도 결정됩니다. 예를 들어, 두 수준의 색조 의미 체계를 추가하면 플롯이 두 개의 선과 오류 대역으로 분할되고 각각의 색상을 지정하여 데이터의 하위 집합을 나타냅니다.

```
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

선 플롯에 스타일 의미를 추가하면 기본적으로 선의 대시 패턴이 변경됩니다.

```
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

그러나 대시와 함께 또는 대신 각 관측에 사용 된 마커로 하위 집합을 식별 할 수 있습니다.

```
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri);
```

산점도와 마찬가지로 여러 의미를 사용하여 선 플롯을 만들 때주의해야합니다. 때로는 정보를 제공하지만 구문 분석 및 해석이 어려울 수도 있습니다. 그러나 하나의 추가 변수에 대한 변경 사항 만 검토하는 경우에도 선의 색상과 스타일을 모두 변경하는 것이 유용 할 수 있습니다. 이렇게하면 흑백으로 인쇄하거나 색맹이있는 사람이 볼 때 플롯에 더 쉽게 액세스 할 수 있습니다.

```
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri);
```

반복 측정 데이터로 작업 할 때 (즉, 여러 번 샘플링 된 단위가있는 경우) 의미론을 통해 구분하지 않고 각 샘플링 단위를 개별적으로 플로팅 할 수도 있습니다. 이렇게하면 범례가 복잡해지지 않습니다.

```
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'"));
```

기본 컬러 맵 및 범례 처리 lineplot()는 색조 의미 체계가 범주 형인지 숫자인지에 따라 달라집니다.

```
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            kind="line", data=dots);
```

hue변수가 숫자 이지만 선형 색상 척도로 제대로 표현되지 않을 수 있습니다 . hue변수 의 수준 이 대수적으로 스케일링 되는 경우가 여기에 해당됩니다 . 목록 또는 사전을 전달하여 각 행에 특정 색상 값을 제공 할 수 있습니다.

```
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            palette=palette,
            kind="line", data=dots);
```

또는 컬러 맵이 정규화되는 방식을 변경할 수 있습니다.

```
from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=.7, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            hue_norm=LogNorm(),
            kind="line",
            data=dots.query("coherence > 0"));
```

세 번째 의미 인 크기는 선의 너비를 변경합니다.

```
sns.relplot(x="time", y="firing_rate",
            size="coherence", style="choice",
            kind="line", data=dots);
```

그동안 size변수는 일반적으로 숫자가 될 것입니다, 그것은 라인의 폭의 범주 형 변수를 매핑 할 수도 있습니다. "두꺼운"선과 "가는"선보다 훨씬 더 많은 것을 구분하기 어렵 기 때문에 그렇게 할 때주의하십시오. 그러나 선이 빈도 변동성이 높으면 대시를 인식하기 어려울 수 있으므로이 경우 다른 너비를 사용하는 것이 더 효과적 일 수 있습니다.

```
sns.relplot(x="time", y="firing_rate",
            hue="coherence", size="choice",
            palette=palette,
            kind="line", data=dots);
```

## 날짜 데이터로 플로팅하기

선 그림은 실제 날짜 및 시간과 관련된 데이터를 시각화하는 데 자주 사용됩니다. 이러한 함수는 원래 형식의 데이터를 기본 matplotlib 함수에 전달하므로 matplotlib의 날짜 형식을 눈금 레이블로 지정하는 기능을 활용할 수 있습니다. 그러나 모든 형식화는 matplotlib 레이어에서 발생해야하며 작동 방식을 확인하려면 matplotlib 문서를 참조해야합니다.

```
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

## 패싯이있는 여러 관계 표시

이 자습서에서는 이러한 함수 가 여러 의미 변수를 한 번에 표시 할 수 있지만 항상 효과적인 것은 아니라는 점을 강조했습니다 . 그러나 두 변수 간의 관계가 하나 이상의 다른 변수에 어떻게 의존하는지 이해하고 싶을 때는 어떨까요?

가장 좋은 방법은 하나 이상의 플롯을 만드는 것입니다. 때문에 relplot()에 기초한다 FacetGrid, 이것은 쉽게 할 수 있습니다. 추가 변수의 영향을 표시하려면 플롯의 의미 적 역할 중 하나에 할당하는 대신이를 사용하여 시각화를 "패싯"합니다. 즉, 여러 축을 만들고 각 축에 데이터의 하위 집합을 플로팅합니다.

```
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips);
```

이 방법으로 두 변수의 영향을 표시 할 수도 있습니다. 하나는 열에 대한 패싯과 다른 하나는 행에 대한 패싯입니다. 그리드에 더 많은 변수를 추가하기 시작하면 그림 크기를 줄일 수 있습니다. 크기 FacetGrid는 각 패싯 의 높이와 가로 세로 비율로 매개 변수화됩니다 .

```
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri);
```

변수의 여러 수준에서 효과를 조사하려는 경우 해당 변수를 열에서 패싯 한 다음 패싯을 행으로 "래핑"하는 것이 좋습니다.

```
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```

종종 "격자"플롯 또는 "소다수"라고하는 이러한 시각화는 눈이 전체 패턴과 해당 패턴의 편차를 쉽게 감지 할 수있는 형식으로 데이터를 제공하기 때문에 매우 효과적입니다. scatterplot()및 relplot()에서 제공하는 유연성을 활용해야하지만 일반적으로 여러 개의 단순한 플롯이 하나의 복잡한 플롯보다 더 효과적이라는 점을 항상 염두에 두십시오.
